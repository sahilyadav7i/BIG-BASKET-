```python
#handeling outliers
```

```python
sns.boxplot(bb['sale_price'])
```

```python
Q1 = bb['sale_price'].quantile(0.25)
print(f"Q1 is {Q1}")

Q3 = bb['sale_price'].quantile(0.75)
print(f"Q3 is {Q3}")
```

```python
IQR = Q3 - Q1
print(f"IQR is {IQR}")
```

```python
lower_bound = Q1 - 1.5 * IQR
print(lower_bound)

upper_bound = Q3 + 1.5 * IQR
print(upper_bound)
```

```python
bb['sale_price'] = np.where(bb['sale_price'] < lower_bound, lower_bound, bb['sale_price'])
bb['sale_price'] = np.where(bb['sale_price'] > upper_bound, upper_bound, bb['sale_price'])
```

```python
sns.boxplot(bb['sale_price'])
plt.show()
```

```python
sns.boxplot(bb['market_price'])
```

```python
Q1 = bb['market_price'].quantile(0.25)
print(f"Q1 is {Q1}")

Q3 = bb['market_price'].quantile(0.75)
print(f"Q3 is {Q3}")
```

```python
IQR = Q3 - Q1
print(f"IQR is {IQR}")
```

```python
lower_bound = Q1 - 1.5 * IQR
print(lower_bound)

upper_bound = Q3 + 1.5 * IQR
print(upper_bound)
```

```python
bb['market_pricev'] = np.where(bb['market_price'] < lower_bound, lower_bound, bb['market_price'])
bb['market_price'] = np.where(bb['market_price'] > upper_bound, upper_bound, bb['market_price'])
```

```python
# visualizations ..
```

```python
category_counts = bb['category'].value_counts()
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='lightpink')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Number of Products in Each Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sub_category_counts = bb['sub_category'].value_counts()
plt.figure(figsize=(12, 6))
sub_category_counts.plot(kind='bar', color='cyan')
plt.xlabel('Sub-Category')
plt.ylabel('Count')
plt.title('Number of Products in Each Sub-Category')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Brand Performance and Market Share
brand_market_share = bb['brand'].value_counts(normalize=True) * 100

# Select top N brands for visualization (adjust N as needed)
top_n_brands = 10
top_brands = brand_market_share.head(top_n_brands)

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_brands, labels=top_brands.index, autopct='%1.1f%%', startangle=90)
plt.title(f'Top {top_n_brands} Brands Market Share')
plt.show()
```

```python
#Sub-category Sales Performance make a bar graph
sub_category_sales = bb.groupby('sub_category')['sale_price'].sum().sort_values(ascending=False)
top_n_sub_categories = 10
top_sub_categories = sub_category_sales.head(top_n_sub_categories)
plt.figure(figsize=(12, 6))
top_sub_categories.plot(kind='bar', color='skyblue')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.title(f'Top {top_n_sub_categories} Sub-categories by Sales')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
category_avg_spending = bb.groupby('category')['sale_price'].mean().sort_values(ascending=False)

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
category_avg_spending.plot(kind='barh', color='lightgreen')
plt.xlabel('Average Spending')
plt.ylabel('Category')
plt.title('Average Spending per Category')
plt.tight_layout()
plt.show()
```

```python
#Brand Popularity make colorful horizintal graph
brand_counts = bb['brand'].value_counts().head(10)  # Adjust the number as needed

plt.figure(figsize=(10, 6))
brand_counts.plot(kind='barh', color='salmon')
plt.xlabel('Number of Products')
plt.ylabel('Brand')
plt.title('Top 10 Brands by Number of Products')
plt.tight_layout()
plt.show()
```

```python
#make a graph on most demanding products
product_counts = bb['product'].value_counts().head(10)  # Adjust the number as needed

plt.figure(figsize=(12, 6))
product_counts.plot(kind='bar', color='lightblue')
plt.xlabel('Product')
plt.ylabel('Number of Sales')
plt.title('Top 10 Most Demanding Products')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
#average selling price of top 20 brands make agraph
top_n_brands = 20
top_brands = bb['brand'].value_counts().head(top_n_brands).index
top_brand_data = bb[bb['brand'].isin(top_brands)]
brand_avg_price = top_brand_data.groupby('brand')['sale_price'].mean().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 6))
brand_avg_price.plot(kind='bar', color='coral')
plt.xlabel('Brand')
plt.ylabel('Average Selling Price')
plt.title('Average Selling Price of Top 20 Brands')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
